```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
url = "https://en.wikipedia.org/wiki/List_of_current_UFC_fighters"
```

```python
search="2%2B2"
headers={'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'}
page = requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(page.text,"lxml")
```

```python
table = soup.find_all(class_="wikitable")[1]
```

```python
#table
```

```python
tbody = table.find("tbody")
```

```python
#tbody
```

```python
data = {
    
    "date": [],
    "name": [],
    "nickname": [],
    "division": [],
    "mma_record": []
}
```

```python
data["mma_record"] = [i.find_all_next("td")[8].text.split() for i in tbody]
data["division"] = [i.find_all_next("td")[5].text.strip() for i in tbody]
#data["nickname"] = [i.find_all_next("td")[3].text.strip() for i in tbody]
data["nickname"] = [i.find_all_next("td")[4].text.strip() for i in tbody] 
data["name"] = [i.find_all_next("td")[2].text.strip() for i in tbody]
data["date"] = [i.find_all_next("td")[0].text.strip() for i in tbody]
```
